```python
import numpy as np

A = np.array([[1, 5, 7, 8, 6.1, 9], [10, 4, 6, 9, 2, 12]])
print('The array A is:')
print(A)

print('A\'s dimension: ', A.ndim)
print('A\'s size: ', A.size)

A = A.astype(np.int32, copy=False)
print('\nChange the type of the array A from \'float\' to \'integer32\'')
print(A)

A[0, 4:6] = [11, 16]
A[1, 2:4] = [8, 10]
print('\nModify through index')
print(A)

print('\nAll the elements that are divisible by 3')
print(A[A % 3 == 0])

B = np.random.rand(5, 5)
print('\nGenerate a 5*5 matrix randomly')
print(B)
Bmax = np.max(B)
Bmin = np.min(B)
C = (B - Bmin) / (Bmax - Bmin)
print('Normalize it')
print(C)

d = np.random.rand(12)
d.sort()
d = d[1: -1]
print('\nA vector from 0 to 1 (not including 0 and 1) of size 10')
print(d)
```

```python
import pandas as pd

s1 = pd.Series([18.5, 20, 24, 18, 15, 14, 16], index = ['周一', '周二', '周三', '周四', '周五', '周六', '周日'])
print('Create a Series')
print(s1)

print('\nTemperature from Tue to Thu')
print(s1[1: 4])

print('\nTemperature greater than 18 ℃')
print(s1[s1 > 18])

df_list = [('cat', 2.5, 1, 'yes'),
           ('cat', 3.0, 3, 'yes'),
           ('snake', 0.5, 2, 'no'),
           ('dog', np.nan, 3, 'yes'),
           ('god', 5.0, 2, 'no'),
           ('cat', 2.0, 3, 'no'),
           ('snake', 4.5, 1, 'no'),
           ('cat', np.nan, 1, 'yes'),
           ('dog', 7.0, 2, 'no'),
           ('dog', 3.0, 1, 'no')]
df_column = ['animal', 'age', 'visits', 'priority']
df_index = list(chr(i) for i in range(97,107))     # Use ASCII code to traverse a to j
df = pd.DataFrame(df_list, columns = df_column, index = df_index)
print('\nCreate a DataFrame')
print(df)

print('\nFetch the \'animal\' and \'age\' columns')
print(df[['animal','age']])

print('\nFetch the \'animal\' and \'age\' columns with index of rows[3, 4, 8]')
print(df.loc[df.index[[3,4,8]],['animal','age']])

print('\nFetch the row with missing \'age\' value')
print(df[df['age'].isnull()])

print('\nThe sum of data in \'visits\' column')
print(df['visits'].sum())

print('\nThe number of each animal')
print(df.value_counts('animal'))
```
