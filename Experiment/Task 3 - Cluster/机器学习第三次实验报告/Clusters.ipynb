```python
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt

# Load data
data1 = pd.read_csv('./aniso.txt', delimiter=' ', names=['x','y'])
data2 = pd.read_csv('./blobs.txt', delimiter=' ', names=['x','y'])
data3 = pd.read_csv('./no_structure.txt', delimiter=' ', names=['x','y'])
data4 = pd.read_csv('./noisy_circles.txt', delimiter=' ', names=['x','y'])
data5 = pd.read_csv('./noisy_moons.txt', delimiter=' ', names=['x','y'])

# Apply K-means clustering and Density-based clustering to the datasets
# Visualize the data by a scatter plot
data1.plot.scatter(x='x', y='y', title='The original data')

# K-means
k_means1 = KMeans(n_clusters=3, max_iter=50, random_state=1)
k_means1.fit(data1)
labels_1_k = pd.DataFrame(k_means1.labels_, columns=['Cluster ID'])
result_1_k = pd.concat((data1,labels_1_k), axis=1)

# Density-based
db1 = DBSCAN(eps=0.5, min_samples=5)
db1.fit(data1)
labels_1_d = pd.DataFrame(db1.labels_, columns=['Cluster ID'])
result_1_d = pd.concat((data1,labels_1_d), axis=1)

# Display
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
result_1_k.plot.scatter(x='x', y='y', c='Cluster ID', colormap='jet', ax=ax1)
ax1.set_title('K-means Clustering')
result_1_d.plot.scatter(x='x', y='y', c='Cluster ID', colormap='jet', ax=ax2)
ax2.set_title('Density-based Clustering')
plt.show()
```
